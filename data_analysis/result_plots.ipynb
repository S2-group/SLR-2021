```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.interpolate as interpolate
```

```python
total_scores = [0, 1, 2, 3, 4]
no_of_studies = [13, 18, 16, 11, 6]

labels = ['Subjects', 'Context', 'Scale', 'Research\nmethod']
score_zero = [26, 27, 33, 50]
score_one = [38, 37, 31, 14]

x = np.arange(len(labels))
width = 0.30  

fig, (ax1, ax2) = plt.subplots(2)

rects = ax1.bar(total_scores, no_of_studies, 1, color='lightgray', edgecolor="black") 
ax1.set_ylabel('Number of studies')
ax1.set_xlabel('Cumulative industrial score')


rects1 = ax2.bar(x - width/2, score_zero, width, label='0', color='gray', edgecolor="black")
rects2 = ax2.bar(x + width/2, score_one, width, label='1', color='lightgray', edgecolor="black")

ax2.set_ylabel('Number of studies')
ax2.set_xticks(x)
ax2.set_xticklabels(labels)
ax2.legend()


def autolabel(rects, ax):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects, ax1)        
autolabel(rects1, ax2)
autolabel(rects2, ax2)

ax1.set_ylim(0, 25)
ax2.set_ylim(0, 70)

fig.tight_layout()

#plt.show()
plt.savefig('figures/industrial_score.pdf', bbox_inches='tight')
```

```python
total_scores = [0, 0.5, 1, 1.5, 2, 2.5, 3]
no_of_studies = [2, 5, 17, 18, 9, 2, 11]

labels = ['Context', 'Study design', 'Validity']
score_low = [2, 8, 46]
score_med = [27, 34, 6]
score_high = [35, 22, 12]

x = np.arange(len(labels))
width = 0.20

fig, (ax1, ax2) = plt.subplots(2)

rects = ax1.bar(total_scores, no_of_studies, 0.5, color='lightgray', edgecolor="black") 
ax1.set_ylabel('Number of studies')
ax1.set_xlabel('Cumulative rigor score')

rects1 = ax2.bar(x - width, score_low, width, label='Low', color='lightgray', edgecolor="black")
rects2 = ax2.bar(x, score_med, width, label='Medium', color='darkgray', edgecolor="black")
rects3 = ax2.bar(x + width, score_high, width, label='High', color='gray', edgecolor="black")

ax2.set_ylabel('Number of studies')
ax2.set_xticks(x)
ax2.set_xticklabels(labels)
ax2.legend()


def autolabel(rects, ax):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects, ax1)        
autolabel(rects1, ax2)
autolabel(rects2, ax2)
autolabel(rects3, ax2)

ax1.set_ylim(0, 25)
ax2.set_ylim(0, 55)
ax2.set_xlim(-0.5, 2.6)

fig.tight_layout()

plt.savefig('figures/rigor_score.pdf', bbox_inches='tight')
```
